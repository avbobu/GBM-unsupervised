```python
accuracy = []
sizes = []
for l in cluster_labels:
    sub_G = G.subgraph([node for node in G.nodes if G.nodes[node]['label'] == l])
    accuracy.append(Spectral_Clustering(sub_G, 2).accuracy)
    sizes.append(sub_G.number_of_nodes())
print(np.average(accuracy, weights = sizes))
```

```python
def sc_iteration(G, cluster):
    G1 = G.subgraph([node for node in G.nodes if G.nodes[node]['label'] == cluster])
    temp = Spectral_Clustering(G1)
    zeros = [G1.nodes[node]["coordinate"] for node in G1.nodes if G1.nodes[node]['label'] == 0]
    ones = [G1.nodes[node]["coordinate"] for node in G1.nodes if G1.nodes[node]['label'] == 1]

    sns.distplot( zeros, color="red", label="Cluster 0")
    sns.distplot( ones, color="skyblue", label="Cluster 1")
    plt.legend()
    
    return (temp,G1)
```
